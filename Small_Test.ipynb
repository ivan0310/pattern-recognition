```python
from sympy import *
num = int(input())#給使用者輸入決定要加到幾
x = symbols('x', communtative = True)#symbols: 定義所用的符號，communtative = True 將符號設定為可交換
f = symbols('f', cls = Function)#cls = Function 是將這些符號設定為函數的代號。
f = x**3 #方程式
result = 0 #儲存結果
for i in range(num+1):
  result = result + f.evalf(subs = {x:i})
print(result)
```
