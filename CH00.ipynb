```python
# n 變數 線性方程組 
n= np.random.randint(2,10)#隨機的2~10數字
print(n) 
np.random.seed(0)
A= np.random.randint(-5,5,size=(n,n))#藉由上面的n給定矩陣的大小
print(A) 
np.random.seed(1) 
b= np.random.randint(-5,5,size=(n,1)) #n*1的矩陣
print(b)
x= np.linalg.solve(A, b) #解方程式
print("x="+str(x)) #顯示結果
print(f''' n= {n},  A= {A} b= {b} x= {x} ''') #顯示結果
```
